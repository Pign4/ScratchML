```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create random line
np.random.seed(3)
beta = np.random.randint(1,6,(2,))
print(beta)
X = np.linspace([1, 0],[1, 10],1000)
y = X @ beta

# plot line
plt.plot(X[:,1],y,color="red")
plt.show()
```

```python
# create noise
mu, sigma = 0, 1 # mean and standard deviation
noise = np.random.normal(mu, sigma, 1000)

# plot noise
plt.hist(noise)
```

```python
# add noise to y
y_noise = y + noise

# plot y_noise
plt.scatter(X[:,1],y_noise, color="blue")
```

```python
# estimate beta in order to minimize the square errors
beta_ = np.linalg.inv(X.T @ X) @ X.T @ y_noise
print(beta_)
```

```python
# estimate y given beta_
y_ = X @ beta_
```

```python
# plot y_ and y_noise
plt.scatter(X[:,1],y_noise,color="blue")
plt.plot(X[:,1],y_,color="orange", lw=3)
```

```python
# compare original and regression line
plt.subplot(121)
plt.plot(X[:,1],y,color="red")
plt.title("Original")

plt.subplot(122)
plt.plot(X[:,1],y_,color="orange")
plt.title("Regression")
```

```python
# checking with built-in functions
np.linalg.lstsq(X, y_noise, rcond=None)[0]
```
